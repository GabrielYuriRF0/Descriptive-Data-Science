## Imports.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly as py
import plotly.graph_objs as go
from sklearn.cluster import KMeans
```

## Carregando o **Dataset**

```python
# Dataset:
# https://drive.google.com/file/d/1Ez7TUtxvBZeRdIUvCT6KPqI9qirmkLSA/view?usp=sharing

df = pd.read_csv('clientes.csv')
df.head()
```

## Primeiro Agrupamento (Cálculo + Plot)

### Definindo a entrada do algoritmo (X)

```python
# código
X = df[['Renda','Pontuacao Consumo']].to_numpy()
plt.scatter(X[:, 0], X[:, 1], s=50);
```

### Cálculo

```python
# código
kmeans = KMeans(n_clusters = 5)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
```

### Plot

```python
# código
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='plasma')

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5);
```
